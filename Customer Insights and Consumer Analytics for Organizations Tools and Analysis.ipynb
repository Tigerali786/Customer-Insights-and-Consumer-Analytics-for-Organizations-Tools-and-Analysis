### Clustering the Customer by Groups

```python
x = pd.DataFrame(df, columns = ['purchase_sum', 'experience'])
```

```python
y = pd.DataFrame(df.characteristic_one)
y.columns = ['Targets']
```

```python
colormap = np.array(['','red','black'])
plt.scatter(x.purchase_sum, x.experience, c=colormap[y.Targets])
```

```python
model = KMeans(n_clusters=2)
```

```python
model.fit(x)
```

```python
model.labels_
```

```python
colormap = np.array(['red', 'black'])
```

```python
plt.scatter(x.purchase_sum, x.experience, c=colormap[model.labels_])
```
